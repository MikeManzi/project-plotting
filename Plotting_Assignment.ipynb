```python
# Plotting histogram that depicts the distribution of Carbon Dioxide
plt.figure(figsize=(12,4))
plt.hist(df['co2'], edgecolor="black", bins=18)
plt.title('Distribution of CO2')
plt.show()
```

```python
# Plotting histogram for city column
plt.figure(figsize=(12,4))
plt.hist(df['city'], edgecolor="black")
plt.title('Histogram of City')
plt.show()
```

```python
# Plotting histogram for feScore column
plt.figure(figsize=(12,4))
plt.hist(df['feScore'], edgecolor="black")
plt.title('Histogram of FE Score')
plt.show()
```

```python
# Plotting the above heatmap


plt.figure(figsize=(6,5))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix)
plt.title('Heatmap')
plt.show()
```

```python
#Data interpretation

#The above heatmap shows the correlation between the columns in the dataset
# Columns like 'comb', 'city', 'UCity','highway', 'UHighway', and 'feScore', show a strong direct proportion to ghgScore (as they increase, the ghgScore also increases).
# Columns like  'cylinders','displ', and 'co2', show a strong inverse proportion to ghgScore (as they increase, the ghgScore decreases).

# We can drop the id, year, pv2 and pv4 columns since there's weak or no correlation to the ghgScore
```
